```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "API_SP.POP.TOTL_DS2_en_csv_v2_12.csv"

# Read the data, skipping metadata rows
df = pd.read_csv(file_path, skiprows=4)

# Select the year for analysis
year = "2020"

# Filter relevant columns
df_filtered = df[["Country Name", year]].dropna()

# Sort values for better visualization
df_filtered = df_filtered.sort_values(by=year, ascending=False).head(20)  # Top 20 most populated countries

# Plot the bar chart
plt.figure(figsize=(12, 6))
plt.bar(df_filtered["Country Name"], df_filtered[year] / 1e6, color="skyblue")
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Population (millions)")
plt.title(f"Top 20 Most Populated Countries in {year}")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```
